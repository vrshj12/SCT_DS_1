```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
```

```python
# Keep only country name + 2024 column
df_clean = df[['Country Name', '2024']].dropna()
```

```python
# Convert population to integers
df_clean['2024'] = df_clean['2024'].astype(int)
```

```python
# Select Top 10 by population
top_countries = df_clean.sort_values(by='2024', ascending=False).head(10)
print(top_countries)
```

```python
# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country Name'], top_countries['2024'], color='teal')
plt.title("Top 10 Most Populous Countries (2024)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
